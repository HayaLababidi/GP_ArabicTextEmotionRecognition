# Read data and character embedding

```
# one hot encoded vectors 



_test_size = 0.33 # percentage of test from the dataset


def read_dataset():
    data_df = pd.read_csv("Emotional-Tone-Dataset.csv", encoding="windows-1256")
    X = data_df[['tweet']].values
    Y = data_df[['label']].values
    
    label_binarizer = LabelBinarizer()
    label_binarizer.fit(Y) # need to be global or remembered to use it later
    one_hot_Y=label_binarizer.transform(Y)
    
    return X, one_hot_Y

def one_hot_encode(dataset):
  max_length = 0
  data_Array2d = []
  
  for i in range(len(dataset)):
    try:
      chars = list(dataset[i][0])
      cur_length = len(dataset[i][0])
      if cur_length > max_length:
        max_length = cur_length
    except:
      print("skipped row{}".format(i))
      continue
    
    length = len(chars)
    
    for indexRow in range(length):
      temp = []
      temp.append(chars[indexRow]) 
      data_Array2d.append(temp) 
      
  print("max_length of sentences",max_length)
  hot = OneHotEncoder(handle_unknown='ignore')
  #print(data_Array2d)
  hot.fit(data_Array2d)
  #print(hot.categories)

  encoded_dataset= [] # list of encoded rows 
  encoded_row = [] # vectors for each char in dataset row 

  # for each row get 2darray of char 
  # encode each char ,append encoded_row
  # append encoded dataset
  
  for row in dataset:
    try:
      chars = list(row[0])
    except:
      print(row[0])
    length = len(chars)
    char_2d = []
    for indexRow in range(length):
      temp = []
      temp.append(chars[indexRow])
      char_2d.append(temp)
    
    encoded_row = hot.transform(char_2d).toarray()
    if len(encoded_dataset) < 3 : 
      x = np.array(encoded_row)
      print("encoded_row shape",x.shape)
      if  encoded_dataset is not None :
        x = np.array(encoded_dataset)
        print("encoded_dataset shape pre append",x.shape)
        
    
    encoded_dataset.append(encoded_row)
    
    if len(encoded_dataset) < 3 : 
      x = np.array(encoded_dataset)
      print("encoded_dataset shape post append",x.shape)
    
  encoded_datasetLength = len(encoded_dataset[0])
  print("encoded_dataset vector lenght", encoded_datasetLength)
  x = np.array(encoded_dataset)
  print("dataset shape",x.shape)
  print("\n------------------")
  return encoded_dataset, max_length, encoded_datasetLength
```

```

#import read_data

def trainModel(X_train, X_test, y_train, y_test, numInput, vectorSize):
    _dropout_rate = 0.2
    _dropout_rate_softmax = 0.5
    _number_of_inputs = numInput #max number of words /characters per doc(tweet)
    _vector_size = vectorSize #vector for each word
    _batch_size = 10
    _kernal_size= 5 #An integer or tuple/list of a single integer
    _pool_size = 3
    _noise_shape = (_batch_size,1,_number_of_inputs)
    _epochs = 100
    _test_size = 0.33 # percentage of test from the dataset
    _Learning_rate = 0.0001


    model = Sequential()

    #input
    # model.add(Input(shape=(_number_of_inputs,_vector_size)))

    #Dropout
    #model.add(keras.layers.Dropout(rate=_dropout_rate,input_shape=(_number_of_inputs,_vector_size)))#,noise_shape,random.randint(0,number_of_inputs)))

    #Convolution
    model.add(Conv1D(filters=_vector_size, kernel_size=_kernal_size,
                     activation="relu", input_shape=(_number_of_inputs,_vector_size)))
    model.add(Conv1D(filters=_vector_size, kernel_size=_kernal_size, activation="relu"))
    model.add(MaxPooling1D(pool_size = _pool_size, padding='valid'))

    #Convolution
    model.add(Conv1D(filters=_vector_size, kernel_size=_kernal_size, activation="relu"))
    model.add(Conv1D(filters=_vector_size, kernel_size=_kernal_size, activation="relu"))
    model.add(MaxPooling1D(pool_size = _pool_size, padding='valid'))

    #Convolution
    model.add(Conv1D(filters=_vector_size, kernel_size=_kernal_size, activation="relu"))
    model.add(Conv1D(filters=_vector_size, kernel_size=_kernal_size, activation="relu"))
    model.add(MaxPooling1D(pool_size = _pool_size, padding='valid'))

    #Convolution
    model.add(Conv1D(filters=_vector_size, kernel_size=_kernal_size, activation="relu"))
    model.add(Conv1D(filters=_vector_size, kernel_size=_kernal_size, activation="relu"))
    model.add(MaxPooling1D(pool_size = _pool_size, padding='valid'))

    #Convolution
    model.add(Conv1D(filters=_vector_size, kernel_size=_kernal_size, activation="relu"))
    model.add(Conv1D(filters=_vector_size, kernel_size=_kernal_size, activation="relu"))
    model.add(MaxPooling1D(pool_size = _pool_size, padding='valid'))


    #Dropout
    model.add(keras.layers.Dropout(_dropout_rate_softmax))
    #,noise_shape,random.randint(0,number_of_inputs)))

    #output
    model.add(keras.layers.Flatten())
    model.add(keras.layers.Dense(8, activation="softmax"))

    opt = keras.optimizers.Adam(lr=_Learning_rate, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False);

    model.compile(
        loss='categorical_crossentropy',
        optimizer=opt,
        metrics=['accuracy']
    )

    tensorboard = TensorBoard(log_dir='./log', histogram_freq=1,
                             write_graph=True,
                             write_grads=True,
                             batch_size=_batch_size,
                             write_images=True)

    model.summary()
    # Train the model
    model.fit(
        X_train,
        y_train,
        batch_size=_batch_size,
        epochs=_epochs,
        validation_data=(X_test, y_test),
        shuffle=True,
        callbacks=[tensorboard]
    )
    model.save("char_embed_mod1.h5")
```
